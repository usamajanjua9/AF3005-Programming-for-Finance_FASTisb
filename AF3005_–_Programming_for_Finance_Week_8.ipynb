## **üìå Key Pandas Objects: Series and DataFrames**  
üîπ **Series**: A **one-dimensional** labeled array (like a list but with indexes).

### **üìç What is a Pandas DataFrame?**  
- A **DataFrame** is like a full spreadsheet table with **multiple columns**.

## **üìå Differences Between Lists, Dictionaries, and DataFrames**
| **Feature** | **List** | **Dictionary** | **Pandas DataFrame** |
|------------|---------|-------------|-----------------|
| **Structure** | Ordered collection | Key-value pairs | Tabular (rows & columns) |
| **Indexing** | Uses numerical indexes | Uses keys | Uses column/row labels |
| **Data Storage** | Stores single type (numbers, strings) | Stores multiple types | Stores structured data |
| **Best for** | Simple lists | Quick key-value lookup | Large financial datasets |

üìå **Example: Storing Financial Data in Different Structures**  
üîπ **List (Simple but unstructured)**  
```python
stock_prices = [175, 2800, 900]  # Only stores values
```

üîπ **Dictionary (Better for lookup but not structured)**  
```python
stock_prices = {"AAPL": 175, "GOOGL": 2800, "TSLA": 900}
```

üîπ **DataFrame (Best for analysis & visualization)**  
```python
import pandas as pd
df = pd.DataFrame({"Stock": ["AAPL", "GOOGL", "TSLA"], "Price": [175, 2800, 900]})
```

‚úî **Pandas DataFrames are the best choice for financial data.**

Indexing Format      | Description                               | Example (on `s = "Python"`) | Output

`s[start:]`         | Extracts from `start` to end            | `s[2:]`                     | `'thon'`

`s[-start:-stop]`  | Negative indexing from right            | `s[-4:-1]`                  | `'yth'`

## **üìå 1. Handling Missing Values in Financial Data**  
üîπ **Missing values** occur in financial datasets due to **data entry errors, incomplete reports, or system issues**.  
üîπ Pandas provides functions to **detect, fill, or remove** missing values.  

üìå **Example: Detecting Missing Values in a Financial Dataset**  
```python
import pandas as pd

# Sample financial dataset
data = {
    "Stock": ["AAPL", "GOOGL", "TSLA", "MSFT"],
    "Price": [175, 2800, None, 310],  # Missing value in TSLA
    "Volume": [50000, None, 60000, 45000]  # Missing value in GOOGL
}

df = pd.DataFrame(data)

# Checking for missing values
print(df.isnull().sum())  # Counts missing values per column
```
**Output:**
```
Stock      0
Price      1
Volume     1
dtype: int64
```
‚úî **Identifies missing values in financial datasets.**  

---

### **üìç 1.1 Filling Missing Values**  
üìå **Fill missing values with a default number (e.g., 0)**  
```python
df.fillna(0, inplace=True)  # Replaces all NaN values with 0
```
‚úî **Useful when missing data should be treated as zero.**  

üìå **Fill missing values with the column mean (useful for stock prices & financial data)**  
```python
df["Price"].fillna(df["Price"].mean(), inplace=True)
```
‚úî **Ensures missing values do not distort financial calculations.**  

---

### **üìç 1.2 Removing Rows with Missing Data**  
üìå **Drop all rows that contain missing values**  
```python
df.dropna(inplace=True)  # Removes rows with missing values
```
‚úî **Used when incomplete data is unreliable.**

## **üìå 1. Filtering and Sorting Stock Market Data**  
üîπ In financial analysis, it is essential to **filter** stock prices based on conditions (e.g., stocks above a certain price) and **sort** them to identify trends (e.g., highest to lowest price).  
üîπ Pandas provides `query()`, `loc[]`, and `sort_values()` for this purpose.  

### **üìç Scenario 1: Finding High-Value Stocks**  
üîπ A stock analyst wants to find all stocks priced above **1000 PKR**.  

üìå **Example: Filtering Stocks Based on Price**  
```python
import pandas as pd

# Sample stock market data
df = pd.DataFrame({
    "Stock": ["AAPL", "GOOGL", "TSLA", "MSFT", "AMZN"],
    "Price": [175, 2800, 900, 310, 3500],
    "Volume": [50000, 80000, 60000, 45000, 70000]
})

# Filter stocks where Price > 1000
high_price_stocks = df[df["Price"] > 1000]
print(high_price_stocks)
```
‚úî **Identifies stocks that are expensive and worth investing in.**  

---

### **üìç Scenario 2: Sorting Stocks by Price (Descending Order)**  
üîπ A trader wants to rank stocks from highest to lowest price to analyze market trends.  

üìå **Example: Sorting Stocks by Price**  
```python
# Sort stocks by price in descending order
sorted_stocks = df.sort_values(by="Price", ascending=False)
print(sorted_stocks)
```
‚úî **Ranks stocks based on price to decide investment strategies.**  

---

### **üìç Scenario 3: Finding Low Liquidity Stocks**  
üîπ A portfolio manager wants to find stocks with **low trading volume** (below **50,000 shares**) to avoid investments in low-liquidity stocks.  

üìå **Example: Filtering Stocks with Low Volume**  
```python
low_liquidity_stocks = df[df["Volume"] < 50000]

```
‚úî **Helps investors avoid stocks with low market activity.**

## **üìå 2. Aggregating Financial Transactions**  
üîπ **Aggregation** is essential in finance to analyze total deposits, withdrawals, and stock price averages.  
üîπ Pandas provides `groupby()` and `agg()` for this purpose.  

### **üìç Scenario 4: Summarizing Deposits and Withdrawals**  
üîπ A bank manager wants to **calculate total deposits and withdrawals** for daily transactions.  

üìå **Example: Grouping Transactions by Type**  
```python
# Sample financial transactions dataset
transactions = pd.DataFrame({
    "Transaction ID": [101, 102, 103, 104, 105],
    "Type": ["Deposit", "Withdrawal", "Deposit", "Withdrawal", "Deposit"],
    "Amount": [5000, 2000, 7000, 1500, 3000]
})

# Group transactions by Type and sum Amount
summary = transactions.groupby("Type")["Amount"].sum()
print(summary)
```
**Output:**
```
Type
Deposit       15000
Withdrawal     3500
```
‚úî **Summarizes deposits and withdrawals to understand cash flow.**

### **üìç Scenario 5: Calculating Average Stock Price Per Sector**  
üîπ A financial analyst wants to analyze stock market trends by **finding the average stock price per sector**.  

üìå **Example: Finding Average Stock Price Per Sector**  
```python
# Sample stock market data
df = pd.DataFrame({
    "Stock": ["AAPL", "GOOGL", "TSLA", "MSFT", "AMZN"],
    "Sector": ["Tech", "Tech", "Auto", "Tech", "E-commerce"],

```
‚úî **Analyzes industry-wide stock performance.**

## **üìå 3. Detecting and Handling Outliers in Financial Data**  
üîπ **Outliers** are unusual data points (e.g., abnormally high withdrawals).  
üîπ Outlier detection helps in **fraud detection and financial risk management**.  

### **üìç Scenario 7: Detecting Suspiciously Large Transactions**  
üîπ A bank wants to detect **unusually large withdrawals** that could indicate fraud.  

üìå **Example: Finding Transactions Above a Threshold**  
```python
# Define an outlier threshold (e.g., transactions above 5000 PKR)
outliers = transactions[transactions["Amount"] > 5000]
print(outliers)
```
‚úî **Flags unusually large transactions for review.**  

---

### **üìç Scenario 8: Removing Outliers from Stock Price Data**  
üîπ A stock market analyst wants to remove **extremely high or low stock prices** that distort analysis.  

üìå **Example: Removing Outliers Based on a Threshold**  
```python
# Define threshold as mean + 2 * standard deviation
threshold = df["Price"].mean() + 2 * df["Price"].std()
df_cleaned = df[df["Price"] < threshold]
print(df_cleaned)
```
‚úî **Ensures accurate financial trend analysis.**

# **Stock Portfolio Analyzer** using **Pandas and Streamlit**, optimized for **Google Colab**.

---

---

### **üìå Expected CSV Format**
To run this project, your CSV file should be structured as follows:

| Stock | Quantity | Buy Price | Current Price |
|--------|---------|-----------|--------------|
| AAPL  | 10      | 175       | 190          |
| TSLA  | 5       | 700       | 650          |
| MSFT  | 8       | 310       | 320          |

---

### **üìå Features of This Project**
‚úî **Users upload a CSV file** containing stock transactions  
‚úî **Calculates portfolio summary** (total investment, current value, gain/loss)  
‚úî **Stock filtering**: Users can select a stock to analyze its performance  
‚úî **Portfolio gain/loss breakdown**: Separates profitable and loss-making stocks  

---

````python
%%writefile Stock_Portfolio_Analyzer.py
# üìå Import necessary libraries
import streamlit as st  # For creating the interactive UI
import pandas as pd  # For handling financial data

# üìå App Title and Description
st.title("üìà Stock Portfolio Analyzer")  # App heading
st.write("Upload your stock transaction CSV file to analyze your investment portfolio.")  # Short description

# üìå File uploader for user to upload CSV file
uploaded_file = st.file_uploader("Upload your CSV file", type=["csv"])

# Check if a file is uploaded
if uploaded_file:
    # üìå Read the CSV file into a Pandas DataFrame
    df = pd.read_csv(uploaded_file)

    # üìå Display the first few rows of the uploaded data
    st.subheader("üìä Uploaded Data Preview")
    st.write(df.head())  # Show first 5 rows of the dataset

    # üìå Portfolio Summary Section
    st.subheader("üìå Portfolio Summary")

    # Check if required columns are present in the dataset
    required_columns = ["Stock", "Quantity", "Buy Price", "Current Price"]
    if all(col in df.columns for col in required_columns):

        # üìå Calculate total investment, current value, and profit/loss
        df["Investment"] = df["Quantity"] * df["Buy Price"]  # Total money spent on each stock
        df["Current Value"] = df["Quantity"] * df["Current Price"]  # Current market value of stocks
        df["Profit/Loss"] = df["Current Value"] - df["Investment"]  # Profit or loss calculation

        # üìå Calculate total portfolio summary
        total_investment = df["Investment"].sum()  # Sum of all investments
        total_value = df["Current Value"].sum()  # Sum of all current values
        total_profit = df["Profit/Loss"].sum()  # Sum of all profits/losses

        # üìå Display overall portfolio summary using Streamlit metrics
        st.metric("üí∞ Total Investment (PKR)", f"{total_investment:,.2f}")  # Show total money invested
        st.metric("üìà Current Portfolio Value (PKR)", f"{total_value:,.2f}")  # Show total portfolio value
        st.metric("üìä Total Gain/Loss (PKR)", f"{total_profit:,.2f}", delta=f"{total_profit:,.2f}")  # Show gain/loss

        # üìå Display performance of individual stocks
        st.subheader("üìå Stock Performance Summary")
        st.write(df[["Stock", "Investment", "Current Value", "Profit/Loss"]])  # Show per-stock performance

        # üìå Allow user to select a stock for analysis
        st.subheader("üîç Analyze Individual Stocks")
        selected_stock = st.selectbox("Choose a stock", df["Stock"].unique())  # Dropdown to select a stock
        stock_data = df[df["Stock"] == selected_stock]  # Filter data for selected stock

        # üìå Display details of selected stock
        st.write(f"üìå **Performance of {selected_stock}**")
        st.write(stock_data[["Quantity", "Buy Price", "Current Price", "Investment", "Current Value", "Profit/Loss"]])

        # üìå Portfolio Gain/Loss Breakdown
        st.subheader("üìä Portfolio Gain/Loss Breakdown")

        # Find stocks that made a profit
        gain_stocks = df[df["Profit/Loss"] > 0]
        # Find stocks that are at a loss
        loss_stocks = df[df["Profit/Loss"] < 0]

        # üìå Display profitable stocks
        st.write("‚úÖ **Profitable Stocks:**")
        st.write(gain_stocks[["Stock", "Profit/Loss"]])

        # üìå Display loss-making stocks
        st.write("‚ùå **Loss-making Stocks:**")
        st.write(loss_stocks[["Stock", "Profit/Loss"]])

    else:
        # Display error if CSV format is incorrect
        st.error("‚ùå CSV file must contain 'Stock', 'Quantity', 'Buy Price', and 'Current Price' columns.")
````
