## **📌 What is a List?**  
🔹 A **list** is a **mutable** (modifiable) collection of elements.  
🔹 Lists can store **multiple data types** (integers, floats, strings, etc.).  
🔹 Lists **preserve order**, meaning items stay in the order they were added.  

📌 **Example:** A list storing **daily expenses**  
```python
expenses = [500, 1200, 300, 700, 450]  # List of expenses in PKR
print(expenses)  # Output: [500, 1200, 300, 700, 450]
```

🚀 **Key Takeaways:**  
✔ **Lists store multiple values in one variable.**  
✔ **They are flexible and allow easy modification.**  

---

## **📌 Creating and Modifying Lists**  
### **1️⃣ Creating a List**  
Lists are defined using **square brackets `[ ]`**.
```python
# List of stock prices
stock_prices = [150, 155, 160, 158, 162]
```
✔ Stores **multiple stock prices** in one variable.  

---

### **2️⃣ Accessing Elements in a List (Indexing & Slicing)**  
Each element in a list has an **index** (starting from `0`).  
```python
print(stock_prices[0])  # Output: 150 (First stock price)
print(stock_prices[-1])  # Output: 162 (Last stock price)
```

✔ **Negative indexing** helps access elements from the end.  

---

### **3️⃣ Modifying Lists**  
Lists are **mutable**, meaning they can be changed.  

📌 **Changing an element in a list**  
```python
stock_prices[1] = 157  # Update second price
print(stock_prices)  # Output: [150, 157, 160, 158, 162]
```

📌 **Adding elements using `append()`**  
```python
stock_prices.append(165)  # Add a new stock price at the end
print(stock_prices)  # Output: [150, 157, 160, 158, 162, 165]
```

📌 **Removing elements using `remove()`**  
```python
stock_prices.remove(158)  # Remove specific stock price
print(stock_prices)  # Output: [150, 157, 160, 162, 165]
```

✔ **Lists allow dynamic modification of financial data.**  

---

## **📌 List Operations**
### **1️⃣ Adding Elements to a List**
- `append()` → Adds an element at the **end**.  
- `insert(index, value)` → Adds an element at a **specific index**.  
```python
stock_prices.insert(2, 159)  # Insert 159 at index 2
print(stock_prices)  # Output: [150, 157, 159, 160, 162, 165]
```

✔ **Useful for inserting new stock prices in historical data.**  

---

### **2️⃣ Removing Elements from a List**
- `remove(value)` → Removes a specific value.  
- `pop(index)` → Removes an element at a specific index.  

```python
stock_prices.pop(3)  # Removes the fourth element
print(stock_prices)  # Output: [150, 157, 159, 162, 165]
```

✔ **Helpful for deleting incorrect or outdated financial records.**  

---

### **3️⃣ Sorting and Filtering Lists**
- `sort()` → Sorts the list in **ascending order**.  
- `sorted()` → Returns a **new sorted list** (without modifying original list).  

📌 **Sorting stock prices in ascending order**  
```python
stock_prices.sort()  # Sort the list
print(stock_prices)  # Output: [150, 157, 159, 162, 165]
```

📌 **Sorting stock prices in descending order**  
```python
stock_prices.sort(reverse=True)
print(stock_prices)  # Output: [165, 162, 159, 157, 150]
```

✔ **Sorting helps in analyzing financial trends.**  

---

## **📌 List Comprehensions – Fast Data Processing**
🔹 **List comprehensions** provide a shorter syntax for **creating lists**.  

📌 **Example: Convert stock prices from PKR to USD**  
```python
stock_prices_pkr = [150, 157, 159, 162, 165]
exchange_rate = 0.0057  # PKR to USD conversion rate

stock_prices_usd = [round(price * exchange_rate, 2) for price in stock_prices_pkr]
print(stock_prices_usd)  # Output: [0.86, 0.89, 0.91, 0.92, 0.94]
```

✔ **List comprehensions are useful for bulk financial calculations.**

## **📌 Finance Use Cases of Lists**
📌 **1️⃣ Storing Stock Prices Over Time**  
```python
daily_prices = [150, 152, 154, 155, 153, 158, 160]
```
✔ Tracks **daily price fluctuations**.  

📌 **2️⃣ Managing Monthly Expenses**  
```python
monthly_expenses = [1200, 2000, 1500, 1800, 2200]
total_expense = sum(monthly_expenses)
print(total_expense)  # Output: 8700 (Total spent in the month)
```
✔ Helps in **budget tracking and expense management**.  

📌 **3️⃣ Storing Customer Transactions**  
```python
transactions = [500, -200, 1000, -150, 700]  # Positive for deposits, negative for withdrawals
net_balance = sum(transactions)
print(net_balance)  # Output: 1850 (Final account balance)
```

## **📌 What is a Tuple?**  
🔹 A **tuple** is an **immutable** (unchangeable) collection of elements.  
🔹 Tuples are **ordered**, meaning elements have a fixed position.  
🔹 Tuples store **multiple data types** (numbers, strings, etc.).  

📌 **Example: A tuple storing fixed tax brackets**  
```python
tax_brackets = (10, 15, 20, 25, 30)  # Tuple of tax percentages
print(tax_brackets)  # Output: (10, 15, 20, 25, 30)
```

🚀 **Key Takeaways:**  
✔ **Tuples store fixed values that should not change.**  
✔ **They use less memory compared to lists.**  

---

## **📌 Creating and Accessing Tuples**  
### **1️⃣ Creating a Tuple**  
Tuples are created using **parentheses `( )`**.  
```python
# Tuple of bank interest rates
interest_rates = (3.5, 4.0, 4.5, 5.0)
```
✔ Stores **fixed financial values** safely.  

---

### **2️⃣ Accessing Tuple Elements (Indexing)**  
Each element in a tuple has an **index** (starting from `0`).  
```python
print(interest_rates[0])  # Output: 3.5 (First interest rate)
print(interest_rates[-1])  # Output: 5.0 (Last interest rate)
```
✔ **Negative indexing** helps access elements from the end.  

---

## **📌 Tuple Operations**  
Unlike lists, **tuples cannot be modified** after creation.  

### **1️⃣ Checking Tuple Length**  
```python
print(len(interest_rates))  # Output: 4
```
✔ **Tuples have a fixed number of elements.**  

---

### **2️⃣ Iterating Over a Tuple**  
Tuples can be looped through using a `for` loop.  
```python
for rate in interest_rates:
    print(f"Interest Rate: {rate}%")
```
✔ **Useful for displaying financial constants**.  

---

### **3️⃣ Tuple Concatenation & Repetition**  
📌 **Concatenation (Joining Tuples)**  
```python
new_rates = interest_rates + (5.5, 6.0)
print(new_rates)  # Output: (3.5, 4.0, 4.5, 5.0, 5.5, 6.0)
```
✔ **Combines multiple financial values**.  

📌 **Repetition (Duplicating Elements)**  
```python
repeat_rates = interest_rates * 2
print(repeat_rates)  # Output: (3.5, 4.0, 4.5, 5.0, 3.5, 4.0, 4.5, 5.0)
```
✔ **Duplicates financial data for simulation purposes**.  

---

## **📌 When to Use Tuples Instead of Lists?**  
✔ **Use tuples when data should not change.**  
✔ **Tuples are faster than lists for fixed data.**  
✔ **Ideal for storing financial constants.**  

📌 **Example: Storing immutable financial data**  
```python
# Tuple of fixed loan terms
loan_terms = (12, 24, 36, 48, 60)  # Loan durations in months
```

## **📌 What is a Dictionary?**  
🔹 A **dictionary** is a collection of key-value pairs.  
🔹 Unlike lists and tuples, dictionaries allow **fast data retrieval** using **keys** instead of index positions.  
🔹 Dictionaries are **mutable**, meaning they can be modified.  

📌 **Example: Storing Customer Account Balances**  
```python
customer_balances = {
    "Ali": 5000,
    "Ayesha": 7000,
    "Usman": 3000
}
print(customer_balances)  # Output: {'Ali': 5000, 'Ayesha': 7000, 'Usman': 3000}
```

🚀 **Key Takeaways:**  
✔ **Dictionaries use key-value pairs for fast data lookup.**  
✔ **Keys must be unique**, while values can be duplicated.  

---

## **📌 Creating and Modifying Dictionaries**  
### **1️⃣ Creating a Dictionary**  
Dictionaries are created using **curly braces `{}`** with **key-value pairs**.  
```python
# Dictionary of stock prices
stock_prices = {
    "AAPL": 175,
    "GOOGL": 2800,
    "MSFT": 310
}
```
✔ **Allows fast lookup of stock prices by symbol.**  

---

### **2️⃣ Accessing Dictionary Elements**  
📌 **Accessing a value using a key**  
```python
print(stock_prices["AAPL"])  # Output: 175
```
✔ **Faster than searching in a list.**  

📌 **Using `.get()` to avoid errors**  
```python
print(stock_prices.get("TSLA", "Stock not found"))  # Output: Stock not found
```
✔ **Returns a default value if key is missing.**  

---

### **3️⃣ Modifying a Dictionary**  
📌 **Adding a new key-value pair**  
```python
stock_prices["TSLA"] = 800
print(stock_prices)  # Output: {'AAPL': 175, 'GOOGL': 2800, 'MSFT': 310, 'TSLA': 800}
```

📌 **Updating an existing value**  
```python
stock_prices["AAPL"] = 180  # Change stock price
print(stock_prices["AAPL"])  # Output: 180
```

📌 **Removing a key-value pair using `del`**  
```python
del stock_prices["GOOGL"]
print(stock_prices)  # Output: {'AAPL': 180, 'MSFT': 310, 'TSLA': 800}
```

✔ **Dictionaries allow easy modification of financial data.**  

---

## **📌 Dictionary Operations**
### **1️⃣ Getting Dictionary Keys & Values**
- `.keys()` → Returns all dictionary keys.  
- `.values()` → Returns all dictionary values.  
- `.items()` → Returns key-value pairs.  

```python
print(stock_prices.keys())  # Output: dict_keys(['AAPL', 'MSFT', 'TSLA'])
print(stock_prices.values())  # Output: dict_values([180, 310, 800])
print(stock_prices.items())  # Output: dict_items([('AAPL', 180), ('MSFT', 310), ('TSLA', 800)])
```

✔ **Useful for iterating through data dynamically.**  

---

### **2️⃣ Iterating Through a Dictionary**
📌 **Looping through keys and values**  
```python
for stock, price in stock_prices.items():
    print(f"{stock}: ${price}")
```
✔ **Efficient way to display financial records.**  

---

### **3️⃣ Merging Dictionaries Using `.update()`**  
```python
new_prices = {"AMZN": 3400, "NFLX": 620}
stock_prices.update(new_prices)
print(stock_prices)
```
✔ **Combines multiple stock price records into one dictionary.**  

---

### **4️⃣ Removing Items Using `.pop()`**  
```python
stock_prices.pop("TSLA")
print(stock_prices)  # Output: {'AAPL': 180, 'MSFT': 310, 'AMZN': 3400, 'NFLX': 620}
```
✔ **Removes a key and returns its value.**  

---

## **📌 Nested Dictionaries for Structured Data**
🔹 **Dictionaries can contain other dictionaries** for **structured data storage**.  

📌 **Example: Storing Bank Customer Information**  
```python
customers = {
    "Ali": {"balance": 5000, "account_type": "Savings"},
    "Ayesha": {"balance": 7000, "account_type": "Checking"}
}
print(customers["Ali"]["balance"])  # Output: 5000
```

```python
account_balances = {"Ali": 5000, "Ayesha": 7000, "Usman": 3000}
print(account_balances["Ali"])  # Output: 5000
```

# **📌 Part 1: Sets in Python**  

## **📌 What is a Set?**  
🔹 A **set** is an **unordered collection** of **unique elements**.  
🔹 Sets **automatically remove duplicates** and allow **fast lookups**.  

📌 **Example: A set storing unique stock symbols**  
```python
stocks = {"AAPL", "GOOGL", "MSFT", "AAPL", "TSLA"}
print(stocks)  # Output: {'AAPL', 'GOOGL', 'MSFT', 'TSLA'}
```
✔ **Duplicate "AAPL" is automatically removed.**  

---

## **📌 Creating and Modifying Sets**  
### **1️⃣ Creating a Set**  
Sets are defined using **curly braces `{}`** or `set()`.  
```python
# Set of stock symbols
stock_symbols = {"AAPL", "GOOGL", "MSFT"}
```
✔ **Stores financial data uniquely.**  

---

### **2️⃣ Adding & Removing Elements in Sets**  
📌 **Adding an element using `add()`**  
```python
stock_symbols.add("TSLA")
print(stock_symbols)  # Output: {'AAPL', 'GOOGL', 'MSFT', 'TSLA'}
```
✔ **Adds a new stock symbol to the set.**  

📌 **Removing an element using `remove()`**  
```python
stock_symbols.remove("MSFT")
print(stock_symbols)  # Output: {'AAPL', 'GOOGL', 'TSLA'}
```
✔ **Deletes a stock symbol.**  

---

## **📌 Set Operations**
### **1️⃣ Union of Sets (`union()`)**
🔹 **Combines elements from two sets (removes duplicates).**  
```python
tech_stocks = {"AAPL", "MSFT"}
finance_stocks = {"JPM", "GS", "MSFT"}
combined = tech_stocks.union(finance_stocks)
print(combined)  # Output: {'AAPL', 'MSFT', 'JPM', 'GS'}
```
✔ **Merges stock data efficiently.**  

---

### **2️⃣ Intersection of Sets (`intersection()`)**
🔹 **Finds common elements between two sets.**  
```python
common_stocks = tech_stocks.intersection(finance_stocks)
print(common_stocks)  # Output: {'MSFT'}
```
✔ **Finds overlapping financial assets.**  

---

### **3️⃣ Difference of Sets (`difference()`)**
🔹 **Finds elements present in one set but not in another.**  
```python
unique_tech_stocks = tech_stocks.difference(finance_stocks)
print(unique_tech_stocks)  # Output: {'AAPL'}
```
✔ **Identifies tech stocks not found in finance.**  

---

## **📌 Finance Use Cases of Sets**
📌 **1️⃣ Removing Duplicate Transactions**  
```python
transactions = {1000, 2000, 3000, 1000, 5000}
print(transactions)  # Output: {1000, 2000, 3000, 5000}
```
✔ **Ensures unique financial transactions.**  

📌 **2️⃣ Finding Common Investors Between Two Stocks**  
```python
apple_investors = {"Alice", "Bob", "Charlie"}
tesla_investors = {"Charlie", "David"}
common = apple_investors.intersection(tesla_investors)
print(common)  # Output: {'Charlie'}
```
✔ **Finds shared investors between different stocks.**

# **📌 Part 2: Queues in Python**  

## **📌 What is a Queue?**  
🔹 A **queue** follows the **FIFO (First In, First Out)** principle.  
🔹 Used when **data should be processed in order**.  

📌 **Example: A queue managing bank transactions**  
```python
from collections import deque (https://www.geeksforgeeks.org/python-collections-module/)

queue = deque(["Deposit - 5000", "Withdrawal - 2000", "Deposit - 1000"])
print(queue)  # Output: deque(['Deposit - 5000', 'Withdrawal - 2000', 'Deposit - 1000'])
```
✔ **Ensures transactions are processed in order.**  

---

## **📌 Implementing Queues in Python**
### **1️⃣ Creating a Queue**  
Queues are implemented using **`collections.deque`** for fast operations.  
```python
from collections import deque (https://www.geeksforgeeks.org/python-collections-module/)

transaction_queue = deque()
```
✔ **Creates an empty transaction queue.**  

---

### **2️⃣ Enqueue Operation (Adding to Queue)**
📌 **Adding transactions using `append()`**  
```python
transaction_queue.append("Deposit - 5000")
transaction_queue.append("Withdrawal - 2000")
print(transaction_queue)  # Output: deque(['Deposit - 5000', 'Withdrawal - 2000'])
```
✔ **Adds transactions to the queue.**  

---

### **3️⃣ Dequeue Operation (Removing from Queue)**
📌 **Processing transactions using `popleft()`**  
```python
transaction = transaction_queue.popleft()
print(transaction)  # Output: 'Deposit - 5000'
print(transaction_queue)  # Output: deque(['Withdrawal - 2000'])
```
✔ **Processes transactions in FIFO order.**

## **🟢 Project Title: Financial Transaction Manager**  


---


### **💡 Objective:**  
Develop a Python-based **Financial Transaction Manager** that:  
✔ Stores **customer account balances** using **dictionaries**  
✔ Tracks **transaction history** using **lists**  
✔ Ensures **unique transactions** using **sets**  
✔ Processes transactions **in order** using **queues**  
✔ Provides **real-time analysis** of **account balances & large transactions**  

---

## **📌 Step 1: Import Required Libraries**
```python
from collections import deque  # Import deque for queue operations (https://www.geeksforgeeks.org/python-collections-module/)
```

✔ **We will use `deque` for transaction processing.**  

---

## **📌 Step 2: Define Data Structures**
```python
# Dictionary to store customer balances
customer_balances = {
    "Ali": 5000,
    "Ayesha": 7000,
    "Usman": 3000
}

# List to store transaction history
transaction_history = []

# Set to store unique transactions
unique_transactions = set()

# Queue to process pending transactions
transaction_queue = deque()
```

✔ **Dictionaries, lists, sets, and queues are initialized for financial management.**  

---

## **📌 Step 3: Define Functions for Transaction Management**  

### **1️⃣ Function to Deposit Money**
```python
def deposit(customer, amount, txn_id):
    """Handles deposit transactions."""
    if txn_id in unique_transactions:
        print("🚫 Duplicate transaction detected!")
        return
    
    if customer in customer_balances:
        customer_balances[customer] += amount
        transaction_history.append((txn_id, customer, "Deposit", amount))
        unique_transactions.add(txn_id)
        print(f"✅ {customer} deposited PKR {amount}. New Balance: PKR {customer_balances[customer]}")
    else:
        print("❌ Customer not found!")
```

✔ **Ensures unique deposits and updates balances securely.**  

---

### **2️⃣ Function to Withdraw Money**
```python
def withdraw(customer, amount, txn_id):
    """Handles withdrawal transactions."""
    if txn_id in unique_transactions:
        print("🚫 Duplicate transaction detected!")
        return
    
    if customer in customer_balances and customer_balances[customer] >= amount:
        customer_balances[customer] -= amount
        transaction_history.append((txn_id, customer, "Withdrawal", amount))
        unique_transactions.add(txn_id)
        print(f"✅ {customer} withdrew PKR {amount}. New Balance: PKR {customer_balances[customer]}")
    else:
        print("❌ Insufficient balance or customer not found!")
```

✔ **Ensures withdrawals are valid and prevents overdrafts.**  

---

### **3️⃣ Function to Process Queued Transactions**
```python
def process_transactions():
    """Processes transactions in FIFO order."""
    while transaction_queue:
        txn = transaction_queue.popleft()  # Retrieve first transaction
        txn_type, customer, amount, txn_id = txn
        
        if txn_type == "Deposit":
            deposit(customer, amount, txn_id)
        elif txn_type == "Withdrawal":
            withdraw(customer, amount, txn_id)
```

✔ **Processes pending transactions in the correct order.**  

---

## **📌 Step 4: Simulate Financial Transactions**  

### **1️⃣ Add Transactions to the Queue**
```python
# Adding transactions to the queue
transaction_queue.append(("Deposit", "Ali", 3000, "TXN1001"))
transaction_queue.append(("Withdrawal", "Ayesha", 2000, "TXN1002"))
transaction_queue.append(("Deposit", "Usman", 5000, "TXN1003"))
```

✔ **Queued transactions ensure fair processing.**  

---

### **2️⃣ Process All Transactions**
```python
process_transactions()
```

✔ **Executes transactions in FIFO order.**

✔ **Keeps a record of all past transactions.**

✔ **Detects and flags transactions exceeding PKR 4000.**
