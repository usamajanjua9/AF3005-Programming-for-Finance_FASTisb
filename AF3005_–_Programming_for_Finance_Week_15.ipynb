## ðŸŸ¢ 4. Introduction to RDDs (Resilient Distributed Datasets)

---

### ðŸ”¹ What are RDDs?

- ðŸ§  **RDD (Resilient Distributed Dataset)** is the **most basic data structure** in Spark.
- ðŸ§  It is an **immutable, distributed collection** of objects across a cluster.
- ðŸ§  RDDs allow **low-level control** over distributed operations.

âœ… **In short**: RDD = Big lists that can be processed on multiple computers at once.

---

### ðŸ”¹ When Should We Use RDDs?

âœ… Use RDDs when:
- You need **fine-grained control** (custom partitioning, low-level transformations).
- You are working with **unstructured data** (logs, raw text files).
- You care about **fault tolerance** and manual optimizations.

---

### ðŸ”¹ Basic RDD Operations

| Operation | What it does | Example |
|:----------|:-------------|:--------|
| **map()** | Transform each element | Multiply all numbers by 2 |
| **filter()** | Keep elements matching a condition | Keep transactions > $1000 |
| **reduce()** | Combine elements together | Sum all transactions |
| **collect()** | Get the RDD as a list | Retrieve results to the driver |

---

### ðŸ”¥ Code Example: Basic RDD Operations

```python
# Start Spark Session
from pyspark.sql import SparkSession
spark = SparkSession.builder.appName("FinanceRDD").getOrCreate()

# Access SparkContext
sc = spark.sparkContext

# Create an RDD
rdd = sc.parallelize([100, 200, 300, 400, 500])

# Map: Multiply each element by 2
rdd2 = rdd.map(lambda x: x * 2)

# Filter: Keep only elements greater than 500
rdd3 = rdd2.filter(lambda x: x > 500)

# Reduce: Sum all elements
total = rdd3.reduce(lambda x, y: x + y)

# Collect results
print("âœ… Doubled RDD:", rdd2.collect())
print("âœ… Filtered RDD (greater than 500):", rdd3.collect())
print("âœ… Sum of filtered elements:", total)
```

âœ… RDDs are **powerful but harder** to work with compared to DataFrames (which we will see next).

## ðŸ”¹ What are DataFrames?

âœ… A **DataFrame** is a **distributed collection of data** organized into **columns**, similar to:
- a **table** in a relational database,
- an **Excel spreadsheet**,
- or a **Pandas DataFrame** (but for BIG DATA!).

âœ… They are **faster**, **easier**, and **more optimized** than RDDs for most tasks.

âœ… **PySpark DataFrame = Easy + Scalable + Powerful** ðŸ“Šâœ¨

---

## ðŸ”¹ Why Use DataFrames Instead of RDDs?

| Feature | RDD | DataFrame |
|:--------|:---:|:---------:|
| ðŸ§  Data Structure | Basic, raw distributed objects | Structured like SQL tables |
| ðŸš€ Speed | Slower | Much faster due to Catalyst optimizer |
| ðŸ“š Ease of Use | Harder (manual transformations) | Easy (SQL-like commands) |
| ðŸ“ˆ Schema Support | No | Yes (automatic data types) |

âœ… **Conclusion**: Always prefer DataFrames unless you have a specific reason to use RDDs!

## ðŸŸ¢ 6. Spark SQL Basics

### ðŸ”¹ What is Spark SQL?

- Spark SQL allows you to **query structured data** inside DataFrames using standard **SQL syntax** (e.g., SELECT, WHERE, GROUP BY).
- You can treat your huge DataFrames just like **tables** in a database.
- Extremely useful for finance teams who already understand SQL (and love Excel tables!)

âœ… **In short**: You can **use SQL to query big financial datasets** in seconds.

````python
# ðŸ“¦ Install PySpark if needed
!pip install pyspark

# ðŸ“˜ Import Libraries
from pyspark.sql import SparkSession
from pyspark.sql.functions import col, rand, round as pyspark_round, avg, lag, date_add, lit
from pyspark.sql.functions import when
from pyspark.sql.types import IntegerType
from pyspark.sql.window import Window
import matplotlib.pyplot as plt
import pandas as pd

# ðŸ“Œ Step 1: Start Spark Session
spark = SparkSession.builder.appName("FinanceStockSimulation").getOrCreate()

# ðŸ“Œ Step 2: Create a Fake Financial Dataset
start_date = "2020-01-01"

# Generate DataFrame with IDs
base_df = spark.range(0, 500).toDF("id")

# ðŸ“Œ Correct the casting of id to INT for date_add()
df = base_df.withColumn("Day", col("id").cast(IntegerType())) \
            .withColumn("Stock",
                        when((col("id") % 3 == 0), "AAPL")
                        .when((col("id") % 3 == 1), "GOOGL")
                        .otherwise("MSFT")) \
            .withColumn("Date", date_add(lit(start_date), col("id").cast(IntegerType()))) \
            .withColumn("Close", pyspark_round(rand(seed=1234) * 300 + 100, 2))

print("âœ… Sample Data Generated!")
df.show(5)

# ðŸ“Œ Step 3: Register DataFrame as SQL Table
df.createOrReplaceTempView("stocks")

# ðŸ“Œ Step 4: Query 1 â€” Find Highest Closing Prices
highest_prices = spark.sql("""
    SELECT Stock, MAX(Close) AS Max_Close
    FROM stocks
    GROUP BY Stock
    ORDER BY Max_Close DESC
""")
print("ðŸ† Highest Closing Prices")
highest_prices.show()

# ðŸ“Œ Step 5: Calculate 5-Day Moving Average
window_spec = Window.partitionBy("Stock").orderBy("Date").rowsBetween(-4, 0)

df = df.withColumn("5Day_MA", avg("Close").over(window_spec))
print(" 5-Day Moving Average Added")
df.select("Date", "Stock", "Close", "5Day_MA").show(10)

# ðŸ“Œ Step 6: Detect Sudden Jumps (>10% Increase)
df = df.withColumn(
    "Prev_Close",
    lag("Close", 1).over(Window.partitionBy("Stock").orderBy("Date"))
)

df = df.withColumn(
    "Change_Pct",
    ((col("Close") - col("Prev_Close")) / col("Prev_Close")) * 100
)

signals_df = df.filter(
    (col("Change_Pct") > 10)
)

print("ðŸš€ Sudden Jumps Detected (>10% rise)")
signals_df.select("Date", "Stock", "Close", "Prev_Close", "Change_Pct").show()

# ðŸ“Œ Step 7: Visualization (Matplotlib for a Stock)
stock_to_plot = "AAPL"  # Choose AAPL, GOOGL, or MSFT

# Convert to Pandas for plotting
plot_data = df.filter(df.Stock == stock_to_plot).select("Date", "Close").toPandas()
plot_data["Date"] = pd.to_datetime(plot_data["Date"])
plot_data.sort_values("Date", inplace=True)

plt.figure(figsize=(10,5))
plt.plot(plot_data["Date"], plot_data["Close"], marker="o")
plt.title(f" {stock_to_plot} Stock Price Simulation")
plt.xlabel("Date")
plt.ylabel("Close Price (USD)")
plt.grid(True)
plt.show()
````
