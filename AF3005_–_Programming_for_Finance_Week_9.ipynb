## **üîπ Creating a Simple Line Plot Using Matplotlib**
A **line chart** is commonly used in finance to **show stock price trends** over time.



### **üìå Example: Stock Price Trends Over 5 Days**
```python
import matplotlib.pyplot as plt

# Sample stock price data
days = ["Mon", "Tue", "Wed", "Thu", "Fri"]
prices = [120, 125, 130, 128, 135]

# Creating a line chart
plt.plot(days, prices, marker='o', linestyle='-', color='b', label="Stock Price")

# Adding titles and labels
plt.title("Stock Price Trend Over the Week")  # Title of the chart
plt.xlabel("Days")  # X-axis label
plt.ylabel("Price (USD)")  # Y-axis label

# Adding a legend
plt.legend()

# Displaying the plot
plt.show()
```
### **üìå Output:**
This will generate a **line plot** showing stock prices over the week.

‚úÖ **Key Features:**
- **`marker='o'`** ‚Üí Adds circular markers to data points.
- **`linestyle='-'`** ‚Üí Connects points with a solid line.
- **`color='b'`** ‚Üí Uses blue for the line color.
- **`plt.legend()`** ‚Üí Displays a legend.


## **üîπ Handling Date-Based Financial Data**
Financial data is often **time-based (dates as X-axis values)**.

### **üìå Example: Plotting a Time-Series Stock Chart**
```python
import matplotlib.dates as mdates
import datetime

# Sample stock data with dates
dates = [datetime.date(2023, 1, i) for i in range(1, 6)]
prices = [120, 125, 130, 128, 135]

plt.figure(figsize=(8,5))  # Set figure size
plt.plot(dates, prices, marker="o", linestyle="-", color="blue")

# Formatting date labels
plt.gca().xaxis.set_major_formatter(mdates.DateFormatter('%b %d'))  
plt.gca().xaxis.set_major_locator(mdates.DayLocator(interval=1))  

plt.xlabel("Date")
plt.ylabel("Stock Price (USD)")
plt.title("Stock Price Over Time")
plt.show()
```
‚úÖ **Key Features:**
- **`mdates.DateFormatter('%b %d')`** ‚Üí Converts dates to readable format (e.g., **Jan 01, Jan 02**).
- **`gca().xaxis.set_major_locator()`** ‚Üí Controls how often dates appear.

---



Run !!

````python
import matplotlib.dates as mdates
import datetime

# Sample stock data with dates
dates = [datetime.date(2023, 1, i) for i in range(1, 6)]
prices = [120, 125, 130, 128, 135]

plt.figure(figsize=(8,5))  # Set figure size
plt.plot(dates, prices, marker="o", linestyle="-", color="blue")

# Formatting date labels
plt.gca().xaxis.set_major_formatter(mdates.DateFormatter('%b %d'))
plt.gca().xaxis.set_major_locator(mdates.DayLocator(interval=1))

plt.xlabel("Date")
plt.ylabel("Stock Price (USD)")
plt.title("Stock Price Over Time")
plt.show()
````

### **üîπ Creating Multiple Plots in One Figure (`subplots`)**
Sometimes, we need to visualize multiple **financial insights in a single figure**. This is useful for **comparing different stocks, revenue vs. expenses, or stock trends over time**.

#### **üìå Example: Plotting Stock Prices for Multiple Companies**
```python
import matplotlib.pyplot as plt

# Sample data for three stocks
days = ["Mon", "Tue", "Wed", "Thu", "Fri"]
stock_A = [120, 125, 130, 128, 135]
stock_B = [115, 118, 125, 127, 132]
stock_C = [90, 95, 100, 102, 110]

# Creating subplots (2 rows, 2 columns)
fig, axes = plt.subplots(2, 2, figsize=(10, 6))

# Line plots for each stock
axes[0, 0].plot(days, stock_A, marker="o", color="blue")
axes[0, 0].set_title("Stock A Prices")

axes[0, 1].plot(days, stock_B, marker="s", color="red")
axes[0, 1].set_title("Stock B Prices")

axes[1, 0].plot(days, stock_C, marker="d", color="green")
axes[1, 0].set_title("Stock C Prices")

# Hiding empty subplot (since we have only 3 stocks)
axes[1, 1].axis("off")

plt.tight_layout()
plt.show()
```

‚úÖ **Key Features:**
- `fig, axes = plt.subplots(2, 2, figsize=(10,6))` ‚Üí Creates a 2x2 grid of plots.  
- `axes[row, col].plot()` ‚Üí Selects each subplot for individual data.  
- `plt.tight_layout()` ‚Üí Adjusts spacing between plots.  

---


Run !!

````python
import matplotlib.pyplot as plt

# Sample data for three stocks
days = ["Mon", "Tue", "Wed", "Thu", "Fri"]
stock_A = [120, 125, 130, 128, 135]
stock_B = [115, 118, 125, 127, 132]
stock_C = [90, 95, 100, 102, 110]

# Creating subplots (2 rows, 2 columns)
fig, axes = plt.subplots(2, 2, figsize=(10, 6))

# Line plots for each stock
axes[0, 0].plot(days, stock_A, marker="o", color="blue")
axes[0, 0].set_title("Stock A Prices")

axes[0, 1].plot(days, stock_B, marker="s", color="red")
axes[0, 1].set_title("Stock B Prices")

axes[1, 0].plot(days, stock_C, marker="d", color="green")
axes[1, 0].set_title("Stock C Prices")

# Hiding empty subplot (since we have only 3 stocks)
axes[1, 1].axis("off")

plt.tight_layout()
plt.show()
````

````python


# Import necessary libraries
import yfinance as yf
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Function to Print Step Headings in Output
def print_heading(text):
    print("\n" + "="*50)
    print(f"üìå {text}")
    print("="*50)

# üìå Step 2: Fetch Stock Market Data from Yahoo Finance
print_heading("Step 2: Fetching Real-World Stock Data")
stocks = ["AAPL", "GOOGL", "TSLA", "AMZN"]  # List of stock tickers

# Download stock data for the last 6 months
stock_data = yf.download(stocks, start="2023-10-01", end="2024-04-01")

# Display first 5 rows
print("üìå Available Stock Data:")
print(stock_data.head())

# üìå Step 3: Debug & Handle Missing Data
print_heading("Step 3: Checking Available Columns & Handling Missing Data")

# Print available columns
print("üìå Available Columns in Dataset:")
print(stock_data.columns)

# Check if dataset is empty
if stock_data.empty:
    raise ValueError("üö® No stock data retrieved! Check stock tickers and date range.")

# üìå Step 4: Extract Adjusted Close Prices
print_heading("Step 4: Extracting Adjusted Close Prices")

# Handle Multi-Level Column Format
if isinstance(stock_data.columns, pd.MultiIndex):
    adj_close = stock_data.xs("Close", axis=1, level=0)  # Extract Close prices
else:
    adj_close = stock_data.get("Close", pd.DataFrame())  # Extract Close prices safely

# Ensure we have valid data
if adj_close.empty:
    raise ValueError("üö® No 'Close' price data found! Check stock tickers and date range.")

# Display first few rows
print("üìå Extracted Adjusted Close Prices:")
print(adj_close.head())

# üìå Step 5: Plot Stock Price Trends Over Time
print_heading("Step 5: Plotting Stock Price Trends Over Time")

plt.figure(figsize=(12,6))
for stock in stocks:
    sns.lineplot(data=adj_close, x=adj_close.index, y=stock, label=stock)
plt.title(" Stock Price Trends Over 6 Months")
plt.xlabel("Date")
plt.ylabel("Stock Price (USD)")
plt.legend()
plt.xticks(rotation=45)
plt.show()

# üìå Step 6: Analyze Stock Volatility
print_heading("Step 6: Analyzing Stock Price Volatility")

plt.figure(figsize=(10,6))
for stock in stocks:
    sns.histplot(adj_close[stock], kde=True, label=stock, alpha=0.5)
plt.title(" Stock Price Distribution & Volatility")
plt.xlabel("Stock Price (USD)")
plt.legend()
plt.show()

# üìå Step 7: Compute & Plot Stock Correlations
print_heading("Step 7: Computing Stock Market Correlation Matrix")

# Compute correlation matrix
corr_matrix = adj_close.corr()

# Create a heatmap
plt.figure(figsize=(8,6))
sns.heatmap(corr_matrix, annot=True, cmap="coolwarm", linewidths=0.5)
plt.title(" Stock Price Correlation Heatmap")
plt.show()

# üìå Step 8: Compute & Plot Daily Stock Returns
print_heading("Step 8: Computing Daily Stock Returns")

# Compute daily percentage change
daily_returns = adj_close.pct_change()

# Display first few rows
print("üìå Daily Stock Returns:")
print(daily_returns.head())

# Plot daily stock returns
plt.figure(figsize=(12,6))
for stock in stocks:
    sns.lineplot(data=daily_returns, x=daily_returns.index, y=stock, label=stock)
plt.title(" Daily Stock Returns Over 6 Months")
plt.xlabel("Date")
plt.ylabel("Daily Return (%)")
plt.legend()
plt.xticks(rotation=45)
plt.show()

# üìå Step 9: Build Financial Dashboard
print_heading("Step 9: Building Financial Dashboard")

fig, axes = plt.subplots(2, 2, figsize=(12, 8))

sns.lineplot(data=adj_close, x=adj_close.index, y="AAPL", ax=axes[0, 0])
axes[0, 0].set_title(" AAPL Stock Price Trend")

sns.histplot(adj_close["TSLA"], kde=True, color="purple", ax=axes[0, 1])
axes[0, 1].set_title(" TSLA Stock Price Distribution")

sns.heatmap(corr_matrix, annot=True, cmap="coolwarm", ax=axes[1, 0])
axes[1, 0].set_title(" Stock Correlation Matrix")

sns.lineplot(data=daily_returns, x=daily_returns.index, y="AMZN", ax=axes[1, 1])
axes[1, 1].set_title("AMZN Daily Returns")

plt.tight_layout()
plt.show()

print_heading("‚úÖ Analysis Complete! Your Stock Market Insights Are Ready! üéØ")
````
