# **ðŸ”¹ 1. Introduction to Control Structures**  

## **ðŸ“Œ What are Control Structures?**  
Control structures in Python **determine the flow of execution** of a program. They allow programmers to:  
âœ… Make **decisions** (`if`, `if-else`, `elif`)  
âœ… **Repeat tasks** (`for`, `while` loops)  

Control structures are crucial in **financial applications** to automate **decision-making processes** such as:  
- Approving or rejecting a **loan application**.  
- Checking if an **investment is profitable**.  
- Monitoring stock prices to **trigger a buy/sell order**.  

---

## **ðŸ“Œ Why are Control Structures Important in Finance?**  
Financial systems often involve **conditional checks and repetitive calculations**, making control structures essential.  

### **ðŸ“ Real-World Examples:**
âœ” **Loan Approval System:**  
- A bank needs to **approve a loan only if an applicant meets income and credit score requirements**.  

âœ” **Investment Decision-Making:**

âœ” **Stock Market Monitoring:**  
- A script must **track stock prices every hour** and **trigger a trade when conditions are met**.  

---

## **ðŸ“Œ Types of Control Structures in Python**  
Python has **three main types** of control structures:  

| **Type**               | **Purpose**                                      | **Example in Finance** |
|------------------------|------------------------------------------------|------------------------|
| **Conditional Statements (`if`, `if-else`, `elif`)** | Allow programs to make **decisions** based on conditions. | Approving a **loan** if income is above a threshold. |
| **Loops (`for`, `while`)** | Execute code **multiple times** until a condition is met. | **Accumulate interest** over multiple months. |
| **Loop Control Statements (`break`, `continue`, `pass`)** | Modify loop behavior. | Stop analyzing stocks once a **target return is achieved**. |

3ï¸âƒ£ **Third line runs**

# **ðŸ”¹ 2. The `if` Statement**  

## **ðŸ“Œ What is the `if` Statement?**  
The `if` statement in Python is used to **execute a block of code only if a specific condition is met**.

ðŸ‘‰ **If the condition is `True`, the code inside the `if` block executes**.  
ðŸ‘‰ **If the condition is `False`, the code inside the `if` block is skipped**.

## **ðŸ“Œ What is the `if-else` Statement?**  
The `if-else` statement is used **when there are two possible outcomes**:  
- **If the condition is `True`, execute the first block of code**.

ðŸ‘‰ **This ensures that one of the two outcomes always runs**.  

---

## **ðŸ“Œ Syntax of the `if-else` Statement**

ðŸš€ **Difference from `if` Statement**:  
- The `if` statement **may not execute any code if the condition is False**.  
- The `if-else` statement **always executes one of the two blocks**.

## **ðŸ”¹ 4. The `elif` Statement**  

## **ðŸ“Œ What is the `elif` Statement?**  
The `elif` statement in Python allows us to **check multiple conditions sequentially**.  
- **If the first condition is False**, it checks the next condition.  
- **If none of the conditions are True, the `else` block executes** (if provided).  
- This is useful for **multi-way decision-making** in financial applications.  

ðŸš€ **Why Use `elif` Instead of Multiple `if` Statements?**  
- **Efficient execution:** Once a `True` condition is found, the remaining `elif` blocks **are not checked**.  
- **Clear structure:** Easier to read compared to multiple `if` statements.  

---

## **ðŸ“Œ Syntax of the `elif` Statement**
```python
if condition1:
    # Code executes if condition1 is True
elif condition2:
    # Code executes if condition2 is True
elif condition3:
    # Code executes if condition3 is True
else:
    # Code executes if no conditions are True
```

## **ðŸ”¹ 5. Nested Conditions**  

## **ðŸ“Œ What are Nested Conditions?**  
Nested conditions occur when we use **one `if` statement inside another `if` statement**.  
- This allows for **multi-level decision-making**.  
- Each **inner condition** is evaluated only if the **outer condition is True**.  
- Commonly used in **financial applications** where multiple factors impact decision-making.  

ðŸš€ **Why Use Nested Conditions?**  
- **Real-world finance scenarios often require multiple factors to be checked together**.  
- **More flexibility** than simple `if-elif-else` statements.  

---

## **ðŸ“Œ Syntax of Nested Conditions**
```python
if condition1:
    if condition2:
        # Code executes if condition1 and condition2 are both True
    else:
        # Code executes if condition1 is True but condition2 is False
else:
    # Code executes if condition1 is False
```
ðŸ”¹ **The second `if` is inside the first `if` block**, meaning it **only runs if the first condition is True**.

ðŸš€ **Why Use Loop Control Statements in Finance?**
